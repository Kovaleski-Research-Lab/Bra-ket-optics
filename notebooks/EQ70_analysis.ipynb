```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
Ls = 2040e-9
Lr = 2040e-9
As = Ls**2
Ar = Lr**2
wavelength = 1550e-9
d = np.linspace(1e-9, 10*wavelength, 1000)
```

```python
dp = wavelength * d / (2 * Ls)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))


ax.plot(d, dp)
ax.set_xlabel("Distance [m]")
ax.set_ylabel("Separation distance [m]")
```
